## Plot Graph

```python
x=df['age']
y=df['trestbps']
plt.scatter(x,y)
plt.xlabel('age')
plt.ylabel('trestbps')
plt.show()
```

```python
data1=df[['age','trestbps']]
data1.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaling=StandardScaler()
scaled_data=scaling.fit_transform(data1)
```

```python
scaled_data[0]
```

```python
from sklearn.cluster import DBSCAN
```

```python
dbscan_cluster=DBSCAN(eps=0.3,min_samples=2,metric='euclidean',algorithm='auto')
dbscan_cluster
```

```python
dbscan_cluster.fit(scaled_data)
dbscan_cluster.labels_
```

```python
x=data1['age']
y=data1['trestbps']
plt.figure(figsize=(10,10))
plt.scatter(x,y,c=dbscan_cluster.labels_)
```
